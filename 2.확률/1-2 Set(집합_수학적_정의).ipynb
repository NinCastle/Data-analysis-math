# 확률의 수학적 정의

## 표본공간과 확률표본
  
확률의 정의 예)  
1. 동전을 한번 던졌다. 동전이 앞면이 나올 것인가 뒷면이 나올 것인가?
2. 플레잉카드(트럼프 카드) 뭉치에서 카드 한장을 선택하였다. 이 카드는 어떤 무늬인가?
3. 동전을 두 번 던져 나올 경우의 수는?
4. 약속 날짜를 정했을 때, 결정된 날짜가 31일 인가?
5. 과일가게에서 손님이 과일을 하나 샀다. 이 과일은 어떤 과일일까?
6. 주식을 샀을 때 내일 가격은 몇 %로 오를까?
7. 회전하는 원반에 화살을 쏘았다. 화살이 박힌 위치의 각도는 기준선에서 몇도 위치인가?
8. 체온을 측정하였다. 체온이 몇도일까?

위 질문의 예들 공통점은 답을 100% 확신할 수 없다는 점이다.  
어떤 문제는 무엇이 답인지 전혀 예측할 수 없는 것도 있고,  
어떤 문제는 어느 정도의 정확도 혹은 범위 내에 있다고 예측 할 수 있는 것도 있다.  
확률론은 이러한 문제가 어떤 답을 가지는지 또 그 답의 신뢰성이 얼마나 되는지 계산하는 정량적인 방법을 제시한다.  
  
이러한 문제들을 확률론적으로 접근하기 위해서는 우선 확률표본(sample)과 표본공간(sample space)을 정의해야 한다.  
  
* **확률표본(probabilistic sample, random sample), 표본(sample) :** 
    * 풀고자 하는 확률적 문제에서 발생(realize)할 수 잇는 하나의 현상,  혹은 선택(sampled)될 수 있는 하나의 경우를 말한다.  
* **표본공간(sample space) :**
    * 가능한 모든 표본의 집합을 말한다. 보통 $\Omega$(대문자 오메가)라는 그리스 문자로 표기한다. 
    * 표본공간을 정의 한다는 것은 우리가 고려한 범위에서 어떤 표본(경우, 현상)이 가능하고 어떤 표본이 가능하지 않은가를 정의 하는 작업이다.

예를 들어)
* 동전 던지기 문제에서는 "앞면(Head)이 나오는 현상" 또는 "뒷면(Tail)이 나오는 결과"가 각각 표본이 될 수 있다.
    * 위 표본들을 각각 기호로 "H", "T"로 표기한다. 
    * 동전이 세로로 서있는 경우는 극히 드문 예이므로 제외하고 표본을 구성한다. 이때는 표본공간이라는 집합이 "H", "T" 두가지 원소(표본)만으로 구성된다.  
* 플레잉 카드 문제에서는 다이아몬드(◆), 하트(♥), 스페이드(♠), 클럽(♣) 기호를 이용하여 표본을 표시할 수 있다.
  
이러한 기호를 사용하면 1번, 2번 문제의 표본공간은 다음과 같이 표기가 가능하다.
$$
\Omega_1 = \{ H,\;\; T\} \\
\Omega_2 = \{ ◆,\; ♥,\; ♠,\; ♣\}  
$$

2. 4번 문제인 약속 날짜 문제에서 결정된 날짜가 31일인지 아닌지 구하는 문제를 확률론적으로 접근했을 때 표본공간 $\Omega_4$를 구하라.

**표본공간**

1. 표본공간은 풀고자하는 문제에 대한 우리의 지식이나 필요성에 따라 달라질 수 있다.  
예를 들어 과일가게에서 손님이 산 과일을 맞추는 문제에서 표본공간은 그 과일과게에서 파는 과일에 따라 달라진다.  
예를 들어 5번 문제인 과일가게 문제에서 사과("A":Apple)와 오렌지("O":Orange)만 파는 과일가게 였다면 표본공간은 다음과 같을 것이다.  
$$
\Omega_5 = \{A,\;O\}
$$
  
2. 표본이 연속적인 숫자인 경우 표본공간이 무한대의 원소를 가질 수 있다.  
예를 들어 6번 문제에서 삼성전자의 주식은 한국거래소의 규정에 따라 다음날 -30% 부터 30%까지 내리거나 오를 수 있다.  
따라서 표본공간은 다음과 같다.
$$
\Omega_6 = \{x\;:\;-30 \le x \le 30\}
$$
상황에 따라 다르지만 위 표본공간은 구간내의 모든 실수를 원소로 가지므로 원소의 갯수가 무한대이다.

## 사건
사건(event)는 표본공간 $\Omega$의 부분집합이다.  
표본공간 중에서의 일부인 표본의 집합을 뜻한다.  
보통 $A, B, C, \ldots$로 대분자로 표기한다.  
![](https://datascienceschool.net/upfiles/0a82370de26442c0aecd1ce172198beb.png)

위에서 예로 든 동전 표본공간에서는 가능한 사건(부분집합)은 다음과 같다.
$$
\begin{aligned}
A &= \{\} = \emptyset \\
B &= \{H\} \\
C &= \{T\} \\
D &= \{H, T\} = \Omega
\end{aligned}
$$

예로 든 $B$라는 사건은 "동전의 앞면이 나오는 경우"를 뜻하고 $D$라는 사건은 "동전의 앞면이 나오거나 뒷면이 나오는 경우"를 뜻한다. "~가 나오는 경우"라고 말할 때 이 "경우"라는 개념이 바로 사건(부분집합)에 해당한다.

```python
# 코드
A = frozenset([])
B = frozenset(['H'])
C = frozenset(['T'])
D = frozenset(['H', 'T'])
set([A, B, C, D])
```

상위의 코드를 set자료형이 아닌 frozenset자료형으로 만든 이유는 dictionaly key로 사용하기 위함이다.

## 확률
확률(probability)이란, 사건(event부분집합)을 입력하면 숫자(확률값)이 출력되는 함수이다.  
확률이라는 함수의 입력값의 집합 즉, 정의역(domain)은 표본공간의 모든 사건(부분집합)의 집합이다.  --시그마 대수--
  
모든 각각의 사건(부분집합)에 어떤 숫자를 할당(allocate)하는 함수가 **확률**이다.  
보통 대문자 알파벳 $P$로 나타낸다.  
확률 $P$는 함수이고 $P(A)$는 $A$라는 사건(부분집합)에 할당된 숫자를 뜻한다.  

$P$함수는 아래와 같은 3가지 규칙을 따라야한다.  
1. 모든 사건에 대해 확률은 실수이고 0 또는 양수 이다.  
$$P(A) \ge 0$$
2. 표본공간(전체집합)이라는 사건(부분집합)에 대한 확률은 1이다.
$$P(\Omega) = 1$$
3. 공통 원소가 없는 두 사건의 합집합의 확률은 각각의 사건의 확률의 합이다.
$$ A \cap B = \emptyset \rightarrow P(A \cup B) = P(A) + P(B)$$  
  
이 세가지를 [콜모고로프의 공리(Kolmogorov's axioms)](https://m.blog.naver.com/PostView.nhn?blogId=skhfdc96&logNo=3134584&proxyReferer=https%3A%2F%2Fwww.google.com%2F)이라고 한다.

![](https://datascienceschool.net/upfiles/8536530a03c0445dab7e5cf5e7a50bc6.png)
예를 들어 표본공간이 $\Omega = \{H, T\}$일 때 사건(부분집합) $\{H\}$에 대한 확률값 $P(\{H\})$는 "라는 표본이 선택될 확률"이고 사건(부분집합) $\{H,T\}$에 대한 확률값 $P(\{H, T\})$는 "$H$ 또는 $T$라는 표본이 선택될 확률"이다. 따라서 확률이라는 함수를 정의 한다는 것은 "무엇이 선택될 확률이 얼마인가"라는 질문에 대한 답을 모든 경우(사건, 부분집합)에 대해 미리 준비해 놓는 것 또는 할당해 놓은 것일라 할 수 있다.

## 확률은 표본이 아닌 사건을 입력을로 가지는 함수

확률에 대해 사람들이 흔히 가지고 있는 잘돈된 지식은 확률이 "표본 하나 하나에 대해 정의되어 있는 숫자"라는 것이다.   
  
즉, 확률은 표본을 입력받아 숫자(확률값)를 출력하는 함수라고 잘 못 알고 있다.  
예를 들어 주사위를 던져 숫자 1이 나타나는 경우에 대해 다음처럼 생각한다.
$$
P(1) = \frac{1}{6}
$$

하지만 확률은 표본이 아닌 사건(부분집합)에 대해 정의하기 때문에, 위 수식은 틀렸다.  
올바른 식은 다음과 같다. 
$$
P(\{1\}) = \frac{1}{6}
$$

$A = \emptyset, B = \{H\}, C = \{T\}, D = \{H, T\}$

```python
P = {A:0, B:0.4, C:0.6, D:1}
P
```

위 예에서 동전의 앞면이 나오는 경우에 대해 할당된 확률값이 0.5가 아닌 0.4에 주목하고,  
콜모고로프의 공리만 지킨다면 각각의 사건(부분집합)에 대한 확률값은 어떤 값도 할당해도 된다.  

위 와 같은 할당법이 콜모고로프의 공리를 지키는지 확인해본다.  
(1) 모든 확률은 0 이상이다.
$$
\begin{aligned}
P(\emptyset) &= 0 \ge 0\\
P(\{H\}) &= 0.4 \ge 0 \\
P(\{T\}) &= 0.6 \ge 0 \\
P(\{H, T\}) &= 1 \ge 0 \\
\end{aligned}
$$
위의 조건을 만족하므로 첫번째 공리를 만족한다.  
  
(2) 전체집합에 대한 확률은 1이다. 
$$
P(\Omega) = P(\{H, T\}) = 1
$$
위의 조건을 만족하므로 두번쨰 공리를 만족한다.  
  
(3) 교집합이 공집합인 사건(부분집합)의 합집합인 사건의 확률은 각 사건(부분집합)의 확률합이다.  
$$
\begin{aligned}
P(\{H\}) &= 0.4 = P(\{H\} \cup \emptyset) = 0.4 + 0 \\
P(\{T\}) &= 0.6 = P(\{T\} \cup \emptyset) = 0.6 + 0 \\
P(\{H, T\}) &= 1 = P(\{H, T\} \cup \emptyset) = 1 + 0 \\
P(\{H, T\}) &= 1 = P(\{H\} \cup \{T\}) = 0.4 + 0.6
\end{aligned}
$$
교집합이 공집합인 어떤 사건을 선택하더라도 합집합의 확률은 각 사건의 확률의 합과 같다.  따라서 세번째 공리를 만족한다.  
  
위의 세가지 공리를 모두 만족시키므로 위 함수는 확률이라고 할 수 있다.

## 주사위 한면의 확률?
우리는 주사위의 어떤 한 면이 나올 확률을 $\frac{1}{6}$이라고 배웠다.  
하지만 이 값은 주사위 확률의 하나의 예에 지나지 않으면 다른 확률도 가능하다.   
실제로 측정하였을 시 주사위의 한면이 정확히 $\frac{1}{6}$로 나오지 않는다.  
그 이유는 주사위의 중심과 단면등 여러가지 것들을 정밀하게 만들어야 하기 때문이다.  
극단적으로 주사위를 조작한다고 해보자.  
조작된 주사위는 절대로 6이 나오지 않으며 10번을 던지면 5번은 반드시 1이 나오게 만들어 져있다.  
이 때 주사위의 확률값이 어떻게 할당되어 있는 상태인가?  
이 때는 확률이 다음과 같이 할당되어 있다고 할 수 있다.
$$
\begin{aligned}
P(\{1\}) &= 0.5\\
P(\{6\}) &= 0
\end{aligned}
$$
  
이 두 사건을 포함한 다른 모든 사건의 확률이 콜모고로프 정의를 만족한다면 이러한 확률값도 확률의 수학적 정의에 맞는다고 할 수 있다.
