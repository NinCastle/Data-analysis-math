# T-검정
회귀분석에서의 t-value(t검정에 대한 통계량)는 독립변수와 종속변수간에 선형관계(관련성)가 존재하는 정도를 나타낸다.  
<br>
How to make T-value

(이성원, 감정평가를 위한 회귀분석의 이해와 활용)

위에서 귀무가설(영가설; null hypothesis)은 두개의 상관관계를(선형관계)가 없다는 것이다.  
그러므로 기울기인 회귀 계수가 0이라는 가정이다.  
그러므로 $\beta_1 = 0$ 이다. 

T검정의 예
<br>


결국 t값은 회귀계수 나누기 표준오차(표준편차)가 된다.  
<br>
회귀계수 나누기 표준오차가 의미하는 것은 무엇인가?  
표준오차가 작아질수록 t값은 커지게 된다.  
표본의 크기(sample 수)가 커질수록 표준오차는 작아진다.  
그러므로 표본의 크기가 커질수록 t값은 커지게 된다.
그래서 표본의 크기가 커질수록 회귀계수에 대한 t값은 유의미하게 나오게 된다.  
<br>
결국 t value는 표준편차가 얼마나 크고 작냐는 이야기를 말한다.  
표준편차의 절대값은 변수의 단위 등에 따라 상대비교가 불가능하므로,  
회귀계수를 표준편차로 나누어 주어서 표준편차의 정도를 상대적으로 비교해 볼수 있게 하는 것이다.
t value가 크다는 것은 표준 편차가 작다는 것이다. (독립-종속 변수간 상관도 높음)  
t value가 작다는 것은 표준 편차가 크다는 것이다. (관계가 낮음)  
그러므로 유의미한 결과를 찾기위해서는 T value가 높아야 한다.
